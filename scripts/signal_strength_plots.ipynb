```python
import numpy as np
from matplotlib import pyplot as plt
import mplhep as hep
```

```python
signal_strengths_exp = {
    r'$m_{T}^{W_2}, \Delta\eta_{jj}$' : '1*0.736*0.752*1.32',
    r'$m_{T}^{W_2}, p_T^{jet_2}$' : '1*0.764*0.728*1.29',
    r'$m_{T}^{W_2}, m_{jj}$' : '1*0.640*0.632*1.48',
    r'$p_T^2, \Delta\eta_{jj}$' : '1*0.522*0.517*1.88',
    r'$p_T^2, p_T^{jet_2}$' : '1*0.518*0.505*1.88',
    r'$p_T^2, m_{jj}$' : '1*0.474*0.488*2.00'
}

signal_strengths_obs = {
    r'$p_T^2, m_{jj}$ combined' : '1.668*0.571*0.569*2.82',
    r'$p_T^2, m_{jj}$ VBF' : '1.404*1.665*1.442*0.86',
    r'$p_T^2, m_{jj}$ ggH' : '1.700*0.627*0.635*2.65'
}
```

```python
plt.style.use(hep.style.CMS)
plt.figure(figsize=(12,10))
for num, (label, s) in enumerate(signal_strengths_exp.items()):
    x, xerrdown, xerrup, s = np.array(s.split('*'), dtype=float)
    plt.errorbar(x, label, xerr = np.array([xerrdown,xerrup]).reshape(2,1), 
                 capsize = 10, capthick = 2, fmt = 'o', markersize = 10, lw=2)
    plt.text(1.9,num,f'${s:.2f}\ \sigma$ (exp.)',va='center')

   
plt.axvline(1, color='k', lw = 2, ls='--')
plt.xlim(0,2.5)
plt.xlabel(r'$\sigma / \sigma_{SM}$ expected', loc='center', fontsize=25)
plt.yticks(fontsize=25)

hep.cms.label('Preliminary', data=True, lumi=59.74, year=2018)
plt.minorticks_off()
plt.tight_layout()
plt.savefig('sig_strength_expected.pdf')
plt.show()
```

```python
plt.style.use(hep.style.CMS)
plt.figure(figsize=(12,10))
for num, (label, s) in enumerate(signal_strengths_obs.items()):
    x, xerrdown, xerrup, s = np.array(s.split('*'), dtype=float)
    
    if x==1.668:
        plt.errorbar(x, label, xerr = np.array([xerrdown,xerrup]).reshape(2,1), 
             capsize = 10, capthick = 2, fmt = 'o', markersize = 10, lw=2, color = 'k')
    else:
        plt.errorbar(x, label, xerr = np.array([xerrdown,xerrup]).reshape(2,1), 
                     capsize = 10, capthick = 2, fmt = 'o', markersize = 10, lw=2)
    plt.text(3,num,f'${s:.2f}\ \sigma$ (obs.)',va="center")

plt.axvline(1, color='k', lw = 2, ls='--')
plt.axvspan(1.668-0.571, 1.668+0.569, alpha=0.1, color='k')
plt.xlim(-1,4.5)
plt.xlabel(r'$\sigma / \sigma_{SM}$ observed', loc='center', fontsize=25)
plt.yticks(fontsize=25)

hep.cms.label('Preliminary', data=True, lumi=59.74, year=2018)
plt.minorticks_off()
plt.tight_layout()
plt.savefig('sig_strength_observed.pdf')
plt.show()
```
